## Popularity Based Recommendor System

```python
# Merging ratings and books with the help of ISBN Column 
Ratings.merge(Books,on ='ISBN')
Ratings.merge(Books,on ='ISBN').shape
```

```python
Ratings_with_name= Ratings.merge(Books,on ='ISBN')
Ratings_with_name
```

```python
#Average rating 
# Finding no. of ratings on each book 
num_rating_df= Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Average rating on books
avg_rating_df= Ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
# Merging num_rating and Ang_rating
popular_df= num_rating_df.merge(avg_rating_df,on = 'Book-Title')
popular_df
```

```python
# keeping only that books whose ratings are above than 250

popular_df[popular_df['num_ratings']>250]
```

## Collaborative Filtering Based Recommender System


```python
# Taking only those users who have done ratings > 200

x = Ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
# These are the users who have done ratings > 200 from the total users 

Ratings_with_name[Ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
Ratings_with_name[Ratings_with_name['User-ID'].isin(padhe_likhe_users)].shape
```

```python
filtered_ratings= Ratings_with_name[Ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_ratings
```

```python
# These are those books which has >50 ratings 

y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
 final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
# These are the final books which has >50 ratings and those users who have done ratings > 200  

pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
# Putting 0 inplace of NaN
pt.fillna(0,inplace=True)
pt
```

```python
# For calculating euclidean distance of books from other books 
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
```

```python
# This recommend function is recommending the books which are similar to the book which is asked through recommend function

def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = Books[Books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
